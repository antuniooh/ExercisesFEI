## Exerc√≠cio 1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fft, ifft
# numero de pontos
N = 64
# numero de pontos em x
T = 100
w=2.0*np.pi/T
t = np.linspace(0.0, T, N)
y = 10*np.cos(25* w*t)
yf = np.fft.fftfreq(N)
masc=yf>0
fft_calculo=np.fft.fft(y)
fft_abs=np.abs(fft_calculo/N)
plt.figure(1)
plt.title("Sinal original")
plt.plot(t,y)
plt.axis([0, 100, -10, 10])
plt.figure(2)
plt.title("Sinal FFT")
plt.plot(yf[masc],fft_abs[masc])
plt.grid()
plt.show()
```

## Exerc√≠cio 2

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fft, ifft
# numero de pontos
N = 1024
# numero de pontos em x
T = 50
w=2.0*np.pi/T
t = np.linspace(0.0, T, N)
y = 10+ 2*np.cos(2*w*t) + 1*np.cos(30*w*t)
yf = np.fft.fftfreq(N)
masc=yf>0
fft_calculo=np.fft.fft(y)
fft_abs=np.abs(fft_calculo/N)
plt.figure(1)
plt.title("Sinal original")
plt.plot(t,y)
plt.axis([0, 50, 0, 20])
plt.figure(2)
plt.title("Sinal FFT")
plt.plot(yf[masc],fft_abs[masc])
plt.grid()
plt.show()
```

```python
from scipy.io import wavfile
from scipy import signal
import matplotlib.pyplot as plt
fs,data=wavfile.read('musica.wav')
nfft=1024
plt.figure(figsize=(7,5))
plt.specgram(data, nfft,fs,noverlap=7*nfft//8)
plt.xlabel('tempo,s')
plt.ylabel('frequencia, Hz')
plt.show()
plt.title('Periodogram de voz')
f, Pxx_den = signal.periodogram(data,fs)
plt.plot(f, Pxx_den)
plt.axis([80, 140, 0, 15])
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.show()
```
